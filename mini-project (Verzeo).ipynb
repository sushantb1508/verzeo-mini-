```python
#taking seed so that every time the random values come out to be constant
cols=['math score', 'reading score',
       'writing score']
np.random.seed(6)
for i in cols:
    sample = np.random.choice(a=dataset[i],size=100)
    #sample mean
    print("sample mean for ",i," :" , sample.mean())
    #population mean
    print ("population mean for ",i," :", dataset[i].mean())
```

```python
#checking the confidence interval or all three scores
from scipy import stats
import math
np.random.seed(10)
for i in cols:
    print("\nFor ",i)
    
    #taking sample 
    sample_size=100
    sample = np.random.choice(a=dataset[i],size=sample_size)
    sample_mean=sample.mean()
    
    #checking the z-critical value
    z_critical = stats.norm.ppf(q=0.95)
    print("z-critical value ",z_critical)
    
    #get the population standard deviation
    population_stddev = dataset[i].std()
    
    #checking the margin error 
    margin_of_error=z_critical * (population_stddev/math.sqrt(sample_mean))
    
    #defining our confidence interval
    confidence_interval = (sample_mean - margin_of_error,sample_mean + margin_of_error)
    print("confidence interval :",confidence_interval)
    print("true mean :",dataset[i].mean())
```
